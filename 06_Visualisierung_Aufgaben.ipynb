```python
# Datensatz erstellen
# Feste Basisdaten pro Produkt (statt Zufallszahlen für Reproduzierbarkeit)
produkte = ["Laptop", "Tablet", "Smartphone", "Smartwatch"]
monate = ["Jan", "Feb", "Mär", "Apr", "Mai", "Jun"]

basis_umsatz = {
    "Laptop":     [15200, 14800, 16100, 15500, 17200, 16800],
    "Tablet":     [ 8100,  7900,  8400,  8200,  8900,  8600],
    "Smartphone": [22300, 21800, 23100, 22700, 24100, 23500],
    "Smartwatch": [ 5100,  4900,  5300,  5200,  5600,  5400],
}

zeilen = []
for produkt in produkte:
    for idx, monat in enumerate(monate):
        zeilen.append({
            "produkt": produkt,
            "monat":   monat,
            "umsatz":  basis_umsatz[produkt][idx]
        })

df_aufgabe = pd.DataFrame(zeilen)
df_aufgabe["monat"] = pd.Categorical(df_aufgabe["monat"], categories=monate, ordered=True)

print(f"Datensatz: {df_aufgabe.shape}")
display(df_aufgabe.head(8))

# Deine Lösung hier:
```
