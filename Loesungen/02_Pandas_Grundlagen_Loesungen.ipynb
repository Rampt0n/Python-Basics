# 6. Aggregationen — Werte zusammenfassen

Aggregationen verdichten viele Werte zu einem (Summe, Durchschnitt, Anzahl, ...).

````python
# Einfache Aggregationen auf einer Spalte
print(f"Summe Gehälter:        {df['gehalt'].sum():>10,.0f} Euro")
print(f"Durchschnittsgehalt:   {df['gehalt'].mean():>10,.0f} Euro")
print(f"Höchstes Gehalt:       {df['gehalt'].max():>10,.0f} Euro")
print(f"Niedrigstes Gehalt:    {df['gehalt'].min():>10,.0f} Euro")
print(f"Median Gehalt:         {df['gehalt'].median():>10,.0f} Euro")
print(f"Anzahl Mitarbeiter:    {df['gehalt'].count():>10}")
````

````python
# Wie viele Mitarbeiter hat jede Abteilung?
df["abteilung"].value_counts()
````

## 6.1 groupby — Aggregationen pro Gruppe

`groupby` ist das Pendant zu "GROUP BY" in SQL. Es berechnet Kennzahlen für jede Gruppe separat.

Frage: Welche Abteilung zahlt im Durchschnitt am meisten?

````python
# Durchschnittsgehalt pro Abteilung
# Hinweis: Mehrere Methoden können direkt hintereinander aufgerufen werden
# (Method Chaining). Das Ergebnis jeder Methode wird sofort als Eingabe
# für die nächste verwendet. In Session1_Extras wird das vertieft.
df.groupby("abteilung")["gehalt"].mean().round(0)
````

````python
# Mehrere Statistiken gleichzeitig mit .agg()
abt_stats = df.groupby("abteilung").agg(
    anzahl=("name", "count"),
    gehalt_avg=("gehalt", "mean"),
    gehalt_min=("gehalt", "min"),
    gehalt_max=("gehalt", "max"),
    erfahrung_avg=("jahre", "mean")
).round(0)

abt_stats
````

````python
# Nach dem Ergebnis sortieren
abt_stats.sort_values("gehalt_avg", ascending=False)
````

# 7. Fehlende Werte (NaN)

In echten Daten fehlen Werte fast immer. pandas stellt `NaN` (Not a Number) für fehlende Werte.

Der Umgang damit ist ein zentrales Thema in der Datentransformation (Notebook 4).

````python
# Beispiel mit fehlenden Werten
daten_mit_luecken = [
    {"name": "Anna",  "gehalt": 52000, "bonus": 3000},
    {"name": "Klaus", "gehalt": 61000, "bonus": None},   # kein Bonus angegeben
    {"name": "Sara",  "gehalt": None,  "bonus": 1500},   # kein Gehalt angegeben
    {"name": "Tom",   "gehalt": 44000, "bonus": None},
    {"name": "Lisa",  "gehalt": 67000, "bonus": 5000},
]

df2 = pd.DataFrame(daten_mit_luecken)
df2
````

````python
# Fehlende Werte erkennen
print("Fehlende Werte pro Spalte:")
print(df2.isnull().sum())
print(f"\nGesamt fehlend: {df2.isnull().sum().sum()}")
````

````python
# Zeilen mit fehlenden Werten anzeigen
print("Zeilen mit mindestens einem fehlenden Wert:")
df2[df2.isnull().any(axis=1)]
````

````python
# Strategien für fehlende Werte:

# Option A: Fehlende Werte mit einem festen Wert auffüllen
df2_gefuellt = df2.copy()
df2_gefuellt["bonus"] = df2_gefuellt["bonus"].fillna(0)  # Kein Bonus = 0

# Option B: Mit dem Mittelwert auffüllen
mittelwert_gehalt = df2["gehalt"].mean()
df2_gefuellt["gehalt"] = df2_gefuellt["gehalt"].fillna(mittelwert_gehalt)

df2_gefuellt
````

````python
# Option C: Zeilen mit fehlenden Werten entfernen
df2_bereinigt = df2.dropna()
print(f"Vor dropna: {len(df2)} Zeilen")
print(f"Nach dropna: {len(df2_bereinigt)} Zeilen")
df2_bereinigt
````

# 8. Spalten umbenennen und Datentypen ändern

Oft kommen Daten mit unhandlichen Spaltennamen oder falschen Datentypen an.

````python
# Beispiel: CSV-Daten mit schlechten Spaltennamen
df_roh = pd.DataFrame({
    "Mitarbeiter Name": ["Anna", "Klaus", "Sara"],
    "Jahres Gehalt EUR": ["52.000", "61.000", "48.000"],  # Zahlen als Text!
    "Eintrittsdatum": ["2019-03-15", "2016-07-01", "2021-09-30"],
})
print(df_roh.dtypes)
df_roh
````

````python
# Spalten umbenennen
df_roh = df_roh.rename(columns={
    "Mitarbeiter Name": "name",
    "Jahres Gehalt EUR": "gehalt",
    "Eintrittsdatum": "eintrittsdatum"
})

# Datentypen korrigieren
df_roh["gehalt"] = df_roh["gehalt"].str.replace(".", "").astype(int)
df_roh["eintrittsdatum"] = pd.to_datetime(df_roh["eintrittsdatum"])

print(df_roh.dtypes)
df_roh
````

````python
# Mit Datums-Daten rechnen
heute = pd.Timestamp.today()
df_roh["betriebszugehoerigkeit_jahre"] = (heute - df_roh["eintrittsdatum"]).dt.days // 365
df_roh
````

# 9. DataFrames zusammenfügen

Oft liegen Daten auf mehrere Tabellen verteilt vor (wie in einer Datenbank). pandas kann sie zusammenführen.

````python
# Tabelle 1: Mitarbeiter-Stammdaten
mitarbeiter = pd.DataFrame({
    "mitarbeiter_id": [1, 2, 3, 4],
    "name": ["Anna", "Klaus", "Sara", "Tom"],
    "abteilungs_id": [10, 20, 10, 30]
})

# Tabelle 2: Abteilungen
abteilungen = pd.DataFrame({
    "abteilungs_id": [10, 20, 30],
    "abteilung_name": ["Vertrieb", "IT", "HR"],
    "standort": ["Hamburg", "Berlin", "München"]
})

print("Mitarbeiter:")
display(mitarbeiter)
print("Abteilungen:")
display(abteilungen)
````

````python
# Merge (JOIN) — wie in SQL: beide Tabellen über gemeinsamen Schlüssel verbinden
kombiniert = pd.merge(mitarbeiter, abteilungen, on="abteilungs_id", how="left")
kombiniert
````

## Abschluss-Übung: Mitarbeiteranalyse

Erstelle eine vollständige Analyse des ursprünglichen Mitarbeiter-DataFrames `df`.

Aufgaben:
1. Erstelle eine neue Spalte `gehaltsstufe` mit den Werten:
   - `"Junior"` für Gehalt < 50.000
   - `"Senior"` für Gehalt zwischen 50.000 und 60.000
   - `"Lead"` für Gehalt > 60.000

2. Zähle, wie viele Mitarbeiter in jede Gehaltsstufe fallen.

3. Gib für jede Abteilung aus, wie viele Mitarbeiter welche Gehaltsstufe haben.


````python
# Frische Kopie des DataFrames
df = pd.DataFrame([
    {"name": "Anna Schmidt",  "abteilung": "Vertrieb",  "gehalt": 52000, "jahre": 5},
    {"name": "Klaus Müller",  "abteilung": "IT",         "gehalt": 61000, "jahre": 8},
    {"name": "Sara Wagner",   "abteilung": "Vertrieb",  "gehalt": 48000, "jahre": 3},
    {"name": "Tom Becker",    "abteilung": "HR",         "gehalt": 44000, "jahre": 6},
    {"name": "Lisa Braun",    "abteilung": "IT",         "gehalt": 67000, "jahre": 11},
    {"name": "Marc Schäfer",  "abteilung": "HR",         "gehalt": 46000, "jahre": 2},
    {"name": "Julia Hoffmann","abteilung": "Vertrieb",  "gehalt": 55000, "jahre": 7},
])
````

````python
# LÖSUNG
# Aufgabe 1: Hilfsfunktion + apply()
def gehaltsstufe(gehalt):
    if gehalt < 50000:
        return "Junior"
    elif gehalt <= 60000:
        return "Senior"
    else:
        return "Lead"

df["gehaltsstufe"] = df["gehalt"].apply(gehaltsstufe)
print("Gehaltsstufen vergeben:")
display(df[["name", "gehalt", "gehaltsstufe"]])

# Aufgabe 2: Anzahl je Stufe
print("\nAnzahl je Gehaltsstufe:")
display(df["gehaltsstufe"].value_counts())

# Aufgabe 3: Abteilung x Gehaltsstufe
print("\nAufteilung Abteilung x Gehaltsstufe:")
display(df.groupby(["abteilung", "gehaltsstufe"])["name"].count().unstack(fill_value=0))
````

### Aufgabe 3 — Groupby & Aggregation

Nutze denselben Mitarbeiter-DataFrame. Berechne pro Stadt:
- Anzahl Mitarbeiter
- Durchschnittsgehalt (auf 0 Stellen gerundet)
- Maximales Gehalt
- Durchschnittliche Betriebszugehörigkeit

Sortiere nach Durchschnittsgehalt absteigend.

````python
# LÖSUNG
stats = mitarbeiter.groupby('stadt').agg(
    anzahl     = ('name',   'count'),
    gehalt_avg = ('gehalt', 'mean'),
    gehalt_max = ('gehalt', 'max'),
    jahre_avg  = ('jahre',  'mean'),
)
stats = stats.round(0)
stats = stats.sort_values('gehalt_avg', ascending=False)
display(stats)
````

### Aufgabe 4 — Merge

Verbinde die beiden DataFrames über `abteilungs_id`.
Berechne anschließend den durchschnittlichen Bonus pro Standort.

````python
# LÖSUNG
import pandas as pd

personal = pd.DataFrame({
    'name':          ['Anna','Klaus','Sara','Tom','Lisa'],
    'abteilungs_id': [10, 20, 10, 30, 20],
    'gehalt':        [52000, 61000, 48000, 44000, 67000],
    'bonus':         [2600, 4900, 2400, 1800, 6700],
})
abteilungen = pd.DataFrame({
    'abteilungs_id': [10, 20, 30],
    'abt_name':      ['Vertrieb', 'IT', 'HR'],
    'standort':      ['Hamburg', 'Berlin', 'München'],
})

kombiniert = pd.merge(personal, abteilungen, on='abteilungs_id')
display(kombiniert)
print('\nDurchschnittsbonus pro Standort:')
display(kombiniert.groupby('standort')['bonus'].mean().round(0))
````
