- Ergebnisse für Stakeholder kommunizieren
- Datenqualitätsprobleme sichtbar machen

| **seaborn** | Schöne statistische Diagramme, einfachere Syntax |
| **plotly** | Interaktive Diagramme (Zoom, Hover) |

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")

# Jupyter-Einstellung: Diagramme direkt im Notebook anzeigen
%matplotlib inline

# Standardgröße für alle Diagramme
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["font.size"] = 11

print("Bibliotheken bereit")
````

````python
# Unsere Beispieldaten
df = pd.DataFrame([
    {"monat": "Jan", "region": "nord", "kategorie": "Elektronik", "umsatz": 12000, "bestellungen": 8},
    {"monat": "Jan", "region": "süd",  "kategorie": "Elektronik", "umsatz": 9500,  "bestellungen": 6},
    {"monat": "Jan", "region": "ost",  "kategorie": "Elektronik", "umsatz": 7800,  "bestellungen": 5},
    {"monat": "Jan", "region": "nord", "kategorie": "Zubehör",    "umsatz": 3200,  "bestellungen": 25},
    {"monat": "Jan", "region": "süd",  "kategorie": "Zubehör",    "umsatz": 2800,  "bestellungen": 20},
    {"monat": "Jan", "region": "ost",  "kategorie": "Zubehör",    "umsatz": 1900,  "bestellungen": 15},
    {"monat": "Feb", "region": "nord", "kategorie": "Elektronik", "umsatz": 15000, "bestellungen": 10},
    {"monat": "Feb", "region": "süd",  "kategorie": "Elektronik", "umsatz": 11000, "bestellungen": 7},
    {"monat": "Feb", "region": "ost",  "kategorie": "Elektronik", "umsatz": 9200,  "bestellungen": 6},
    {"monat": "Feb", "region": "nord", "kategorie": "Zubehör",    "umsatz": 4100,  "bestellungen": 32},
    {"monat": "Feb", "region": "süd",  "kategorie": "Zubehör",    "umsatz": 3300,  "bestellungen": 24},
    {"monat": "Feb", "region": "ost",  "kategorie": "Zubehör",    "umsatz": 2400,  "bestellungen": 18},
    {"monat": "Mär", "region": "nord", "kategorie": "Elektronik", "umsatz": 13500, "bestellungen": 9},
    {"monat": "Mär", "region": "süd",  "kategorie": "Elektronik", "umsatz": 10200, "bestellungen": 7},
    {"monat": "Mär", "region": "ost",  "kategorie": "Elektronik", "umsatz": 8800,  "bestellungen": 6},
    {"monat": "Mär", "region": "nord", "kategorie": "Zubehör",    "umsatz": 3800,  "bestellungen": 29},
    {"monat": "Mär", "region": "süd",  "kategorie": "Zubehör",    "umsatz": 3100,  "bestellungen": 23},
    {"monat": "Mär", "region": "ost",  "kategorie": "Zubehör",    "umsatz": 2100,  "bestellungen": 17},
])

print(f"Datensatz: {df.shape}")
df.head()
````

---
# 1. Balkendiagramm (Bar Chart)

**Wann?** Kategorien vergleichen (z.B. Umsatz pro Region)

````python
# Daten aggregieren
umsatz_region = df.groupby("region")["umsatz"].sum().sort_values(ascending=False)

# Diagramm erstellen
fig, ax = plt.subplots()  # fig = Zeichenfläche, ax = Achsensystem

bars = ax.bar(
    umsatz_region.index,     # x-Achse: Kategorien
    umsatz_region.values,    # y-Achse: Werte
    color=["#2196F3", "#4CAF50", "#FF9800"],  # Farben
    edgecolor="white",
    linewidth=0.8
)

# Werte über den Balken
for bar in bars:
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # x-Position: Mitte des Balkens
        bar.get_height() + 200,              # y-Position: leicht über dem Balken
        f"{bar.get_height():,.0f} €",        # Text
        ha="center", va="bottom", fontsize=10
    )

# Beschriftungen und Formatierung
ax.set_title("Gesamtumsatz nach Region (Q1 2024)", fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("Region")
ax.set_ylabel("Umsatz (Euro)")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f} €"))
ax.set_ylim(0, umsatz_region.max() * 1.15)

plt.tight_layout()
plt.savefig("output/umsatz_region.png", dpi=150, bbox_inches="tight")
plt.show()
print("Gespeichert: output/umsatz_region.png")
````

---
# 2. Gruppierte Balken und horizontale Balken

````python
# Gruppiertes Balkendiagramm: Umsatz nach Region UND Kategorie
pivot_daten = df.groupby(["region", "kategorie"])["umsatz"].sum().unstack()

fig, ax = plt.subplots()
pivot_daten.plot(kind="bar", ax=ax, width=0.7)

ax.set_title("Umsatz nach Region und Kategorie", fontsize=14, fontweight="bold")
ax.set_xlabel("Region")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(title="Kategorie")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)  # Beschriftung horizontal
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k €"))

plt.tight_layout()
plt.show()
````

````python
# Gestapelte Balken (Stacked Bar) — für Anteile
fig, ax = plt.subplots()
pivot_daten.plot(kind="bar", stacked=True, ax=ax, width=0.6)

ax.set_title("Umsatz-Zusammensetzung nach Region", fontsize=14, fontweight="bold")
ax.set_xlabel("Region")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(title="Kategorie", loc="upper right")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

plt.tight_layout()
plt.show()
````

---
# 3. Liniendiagramm (Line Chart)

**Wann?** Entwicklung über Zeit zeigen

````python
# Umsatz-Trend pro Monat und Region
monat_region = df.groupby(["monat", "region"])["umsatz"].sum().reset_index()

# Monatsreihenfolge festlegen
monat_reihenfolge = ["Jan", "Feb", "Mär"]
monat_region["monat"] = pd.Categorical(monat_region["monat"], categories=monat_reihenfolge, ordered=True)
monat_region = monat_region.sort_values("monat")

fig, ax = plt.subplots()

farben = {"nord": "#2196F3", "süd": "#4CAF50", "ost": "#FF9800"}
marker = {"nord": "o", "süd": "s", "ost": "^"}

for region in ["nord", "süd", "ost"]:
    region_df = monat_region[monat_region["region"] == region]
    ax.plot(
        region_df["monat"],
        region_df["umsatz"],
        marker=marker[region],
        linewidth=2,
        markersize=8,
        label=f"Region {region.capitalize()}",
        color=farben[region]
    )

ax.set_title("Umsatz-Entwicklung nach Region (Q1 2024)", fontsize=14, fontweight="bold")
ax.set_xlabel("Monat")
ax.set_ylabel("Umsatz (Euro)")
ax.legend()
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k €"))
ax.grid(True, alpha=0.4)

plt.tight_layout()
plt.show()
````

# 4. Kreisdiagramm (Pie Chart)

**Wann?** Anteile am Ganzen zeigen (max. 5–6 Segmente)

````python
umsatz_kat = df.groupby("kategorie")["umsatz"].sum()

fig, ax = plt.subplots(figsize=(8, 6))

wedges, texts, autotexts = ax.pie(
    umsatz_kat.values,
    labels=umsatz_kat.index,
    autopct="%1.1f%%",          # Prozentangabe
    startangle=90,               # Startwinkel
    explode=[0.05] * len(umsatz_kat),  # Segmente leicht herausziehen
    colors=["#2196F3", "#4CAF50"],
    shadow=True
)

# Schriftgröße anpassen
for text in autotexts:
    text.set_fontsize(12)
    text.set_fontweight("bold")

ax.set_title("Umsatzanteil nach Kategorie", fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
````

---
# 5. Streudiagramm (Scatter Plot)

**Wann?** Zusammenhänge zwischen zwei Variablen untersuchen

````python
fig, ax = plt.subplots()

# Farben nach Kategorie
farben_map = {"Elektronik": "#2196F3", "Zubehör": "#FF9800"}

for kat, gruppe in df.groupby("kategorie"):
    ax.scatter(
        gruppe["bestellungen"],
        gruppe["umsatz"],
        label=kat,
        color=farben_map[kat],
        s=80,            # Punktgröße
        alpha=0.8,       # Transparenz
        edgecolors="white",
        linewidth=0.5
    )

ax.set_title("Bestellungen vs. Umsatz", fontsize=14, fontweight="bold")
ax.set_xlabel("Anzahl Bestellungen")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(title="Kategorie")
ax.grid(True, alpha=0.3)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k €"))

plt.tight_layout()
plt.show()
````

# 6. Histogramm und Box Plot

**Wann?** Verteilung einer Variablen verstehen

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Histogramm
ax1.hist(df["umsatz"], bins=10, color="#2196F3", edgecolor="white", linewidth=0.8)
ax1.axvline(df["umsatz"].mean(), color="red", linestyle="--", label=f"Mittelwert: {df['umsatz'].mean():,.0f} €")
ax1.axvline(df["umsatz"].median(), color="orange", linestyle="--", label=f"Median: {df['umsatz'].median():,.0f} €")
ax1.set_title("Verteilung der Umsätze", fontsize=13, fontweight="bold")
ax1.set_xlabel("Umsatz (Euro)")
ax1.set_ylabel("Häufigkeit")
ax1.legend()

# Box Plot: Zeigt Median, Quartile, Ausreißer
daten_nach_region = []
for r in ["nord", "süd", "ost"]:
    daten_nach_region.append(df[df["region"] == r]["umsatz"].values)
bp = ax2.boxplot(
    daten_nach_region,
    labels=["Nord", "Süd", "Ost"],
    patch_artist=True,
    boxprops=dict(facecolor="#E3F2FD")
)
ax2.set_title("Umsatz-Verteilung nach Region", fontsize=13, fontweight="bold")
ax2.set_ylabel("Umsatz (Euro)")
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k €"))

plt.tight_layout()
plt.show()
````

# 7. Mehrere Diagramme in einem Layout (Subplots)

````python
# Dashboard-ähnliches Layout mit 4 Diagrammen
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Verkaufs-Dashboard Q1 2024", fontsize=16, fontweight="bold", y=1.02)

# --- Diagramm 1 (oben links): Umsatz nach Region ---
ax = axes[0, 0]
region_sum = df.groupby("region")["umsatz"].sum().sort_values(ascending=True)
ax.barh(region_sum.index, region_sum.values, color="#2196F3")
ax.set_title("Gesamtumsatz nach Region", fontweight="bold")
ax.set_xlabel("Euro")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k"))
for i, v in enumerate(region_sum.values):
    ax.text(v + 300, i, f"{v:,.0f} €", va="center", fontsize=9)

# --- Diagramm 2 (oben rechts): Monatstrend ---
ax = axes[0, 1]
monat_sum = df.groupby("monat")["umsatz"].sum()
monat_sum = monat_sum.reindex(["Jan", "Feb", "Mär"])
ax.plot(monat_sum.index, monat_sum.values, "o-", color="#4CAF50", linewidth=2.5, markersize=8)
ax.fill_between(monat_sum.index, monat_sum.values, alpha=0.2, color="#4CAF50")
ax.set_title("Umsatz-Trend", fontweight="bold")
ax.set_ylabel("Euro")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k €"))
ax.grid(True, alpha=0.3)

# --- Diagramm 3 (unten links): Kategorie-Anteile ---
ax = axes[1, 0]
kat_sum = df.groupby("kategorie")["umsatz"].sum()
ax.pie(kat_sum.values, labels=kat_sum.index, autopct="%1.0f%%",
       colors=["#2196F3", "#FF9800"], startangle=90)
ax.set_title("Umsatz nach Kategorie", fontweight="bold")

# --- Diagramm 4 (unten rechts): Bestellungen nach Monat und Region ---
ax = axes[1, 1]
pivot_best = df.groupby(["monat", "region"])["bestellungen"].sum().unstack()
pivot_best = pivot_best.reindex(["Jan", "Feb", "Mär"])
pivot_best.plot(kind="bar", ax=ax, width=0.7)
ax.set_title("Bestellungen nach Monat und Region", fontweight="bold")
ax.set_xlabel("")
ax.set_ylabel("Anzahl")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.legend(title="Region")

plt.tight_layout()
plt.savefig("output/dashboard.png", dpi=150, bbox_inches="tight")
plt.show()
print("Dashboard gespeichert: output/dashboard.png")
````

Anforderungen:

- Legende

````python
# LÖSUNG
monat_kat = df.groupby(["monat", "kategorie"])["bestellungen"].sum().reset_index()
monat_kat["monat"] = pd.Categorical(monat_kat["monat"], categories=["Jan", "Feb", "Mär"], ordered=True)
monat_kat = monat_kat.sort_values("monat")

fig, ax = plt.subplots(figsize=(10, 6))

farben = {"Elektronik": "#2196F3", "Zubehör": "#FF9800"}
for kat in monat_kat["kategorie"].unique():
    daten = monat_kat[monat_kat["kategorie"] == kat]
    ax.plot(daten["monat"], daten["bestellungen"],
            marker="o", linewidth=2.5, markersize=9,
            label=kat, color=farben[kat])
    # Werte an den Punkten
    for _, row in daten.iterrows():
        ax.annotate(str(int(row["bestellungen"])),
                    (row["monat"], row["bestellungen"]),
                    textcoords="offset points", xytext=(0, 10),
                    ha="center", fontsize=9)

ax.set_title("Bestellungen pro Monat nach Kategorie", fontsize=14, fontweight="bold")
ax.set_xlabel("Monat")
ax.set_ylabel("Anzahl Bestellungen")
ax.legend(title="Kategorie")
ax.grid(True, alpha=0.3)
ax.set_ylim(0, monat_kat["bestellungen"].max() * 1.2)

plt.tight_layout()
plt.savefig("output/bestellungen_trend.png", dpi=150, bbox_inches="tight")
plt.show()
print("Gespeichert!")
````

# 8. Seaborn — Statistische Diagramme

Seaborn vereinfacht viele häufige Visualisierungsaufgaben erheblich.

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Heatmap: Umsatz nach Monat und Region
pivot = df.groupby(["monat", "region"])["umsatz"].sum().unstack()
pivot = pivot.reindex(["Jan", "Feb", "Mär"])

sns.heatmap(
    pivot,
    ax=axes[0],
    annot=True,             # Werte in den Zellen anzeigen
    fmt=",.0f",             # Zahlenformat
    cmap="Blues",           # Farbschema
    linewidths=0.5
)
axes[0].set_title("Umsatz-Heatmap (Monat × Region)", fontweight="bold")

# Violinplot: Verteilung nach Kategorie
sns.violinplot(
    data=df, x="kategorie", y="umsatz",
    ax=axes[1],
    palette="muted", inner="box"
)
axes[1].set_title("Umsatz-Verteilung nach Kategorie", fontweight="bold")
axes[1].set_ylabel("Umsatz (Euro)")

plt.tight_layout()
plt.show()
````

# 9. Interaktive Diagramme mit plotly

Plotly-Diagramme sind interaktiv: Zoom, Hover, Pan — ideal für Reports.

````python
import plotly.express as px

# Interaktives Balkendiagramm
df_agg = df.groupby(["monat", "region"])["umsatz"].sum().reset_index()
df_agg["monat"] = pd.Categorical(df_agg["monat"], categories=["Jan", "Feb", "Mär"], ordered=True)
df_agg = df_agg.sort_values("monat")

fig = px.bar(
    df_agg,
    x="monat",
    y="umsatz",
    color="region",
    barmode="group",
    title="Umsatz nach Monat und Region (interaktiv)",
    labels={"umsatz": "Umsatz (Euro)", "monat": "Monat", "region": "Region"},
    color_discrete_map={"nord": "#2196F3", "süd": "#4CAF50", "ost": "#FF9800"}
)
fig.update_layout(legend_title="Region")
fig.show()
````

# Abschluss-Übung: Vollständige Analyse mit Visualisierung

Kombiniere alles aus den Notebooks 1–6! Erstelle eine vollständige Analyse:

1. Erstelle einen realistischen Datensatz mit Monatsumsätzen für 4 Produkte über 6 Monate.
2. Berechne aggregierte Kennzahlen (Gesamt, Durchschnitt, Trend).
3. Erstelle ein **Dashboard** mit 3 Diagrammen:
   - Umsatz-Trend (Linie) pro Produkt über die Monate
   - Gesamtumsatz pro Produkt (Balken)
   - Marktanteil pro Produkt (Torte)
4. Speichere das Dashboard als PNG.

````python
# Datensatz erstellen (identisch mit Hauptnotebook)
import matplotlib.pyplot as plt

produkte = ["Laptop", "Tablet", "Smartphone", "Smartwatch"]
monate = ["Jan", "Feb", "Mär", "Apr", "Mai", "Jun"]

basis_umsatz = {
    "Laptop":     [15200, 14800, 16100, 15500, 17200, 16800],
    "Tablet":     [ 8100,  7900,  8400,  8200,  8900,  8600],
    "Smartphone": [22300, 21800, 23100, 22700, 24100, 23500],
    "Smartwatch": [ 5100,  4900,  5300,  5200,  5600,  5400],
}

zeilen = []
for produkt in produkte:
    for idx, monat in enumerate(monate):
        zeilen.append({
            "produkt": produkt,
            "monat":   monat,
            "umsatz":  basis_umsatz[produkt][idx]
        })

df_aufgabe = pd.DataFrame(zeilen)
df_aufgabe["monat"] = pd.Categorical(df_aufgabe["monat"], categories=monate, ordered=True)
````

````python
# LÖSUNG
# Aggregationen
gesamt_pro_produkt = df_aufgabe.groupby("produkt")["umsatz"].sum()
trend = df_aufgabe.sort_values("monat")

# Farbliste für Produkte (in Reihenfolge des Index)
farben_liste = []
for p in gesamt_pro_produkt.index:
    farben_liste.append(farben_prod[p])

# Dashboard
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Produkt-Analyse H1 2024", fontsize=16, fontweight="bold")

farben_prod = {"Laptop": "#2196F3", "Tablet": "#4CAF50", "Smartphone": "#E91E63", "Smartwatch": "#FF9800"}

# 1. Liniendiagramm: Trend
ax = axes[0]
for prod in produkte:
    d = trend[trend["produkt"] == prod]
    ax.plot(d["monat"], d["umsatz"], marker="o", linewidth=2, markersize=5,
            label=prod, color=farben_prod[prod])
ax.set_title("Umsatz-Trend", fontweight="bold")
ax.set_xlabel("Monat")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(fontsize=8)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k"))
ax.grid(True, alpha=0.3)
ax.set_xticklabels(monate, rotation=45)

# 2. Balkendiagramm: Gesamt
ax = axes[1]
bars = ax.bar(gesamt_pro_produkt.index, gesamt_pro_produkt.values,
              color=farben_liste)
for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 200,
            f"{bar.get_height()/1000:.0f}k", ha="center", fontsize=8)
ax.set_title("Gesamtumsatz", fontweight="bold")
ax.set_ylabel("Euro")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x/1000:.0f}k"))
ax.set_xticklabels(gesamt_pro_produkt.index, rotation=15)

# 3. Kreisdiagramm: Marktanteil
ax = axes[2]
ax.pie(gesamt_pro_produkt.values, labels=gesamt_pro_produkt.index,
       autopct="%1.0f%%", startangle=90,
       colors=farben_liste)
ax.set_title("Marktanteil", fontweight="bold")

plt.tight_layout()
plt.savefig("output/produkt_dashboard.png", dpi=150, bbox_inches="tight")
plt.show()
print("Dashboard gespeichert!")
````

# Welcher Diagrammtyp für welche Frage?

| Frage | Diagrammtyp | matplotlib |
|-------|-------------|------------|
| Kategorien vergleichen | Balken | `ax.bar()` / `ax.barh()` |
| Entwicklung über Zeit | Linie | `ax.plot()` |
| Anteile am Ganzen | Torte | `ax.pie()` |
| Zusammenhang zweier Variablen | Streudiagramm | `ax.scatter()` |
| Verteilung einer Variable | Histogramm | `ax.hist()` |
| Verteilung & Ausreißer | Box Plot | `ax.boxplot()` |
| Korrelationsmatrix | Heatmap | `sns.heatmap()` |
| Teile + Zeit kombiniert | Gestapeltes Balken | `df.plot(stacked=True)` |

````python
# LÖSUNG
import matplotlib.pyplot as plt, matplotlib.ticker as mticker
import pandas as pd

df_p = pd.DataFrame({'produkt': ['Laptop','Tablet','Smartphone','Smartwatch','Kopfhörer','Kamera'],
                     'umsatz':  [206000, 84000, 148000, 52000, 37000, 95000]})
df_p = df_p.sort_values('umsatz')

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(df_p['produkt'], df_p['umsatz'], color='#2196F3')
for bar in bars:
    ax.text(bar.get_width() + 2000, bar.get_y() + bar.get_height()/2,
            f"{bar.get_width()/1000:.0f}k €", va='center', fontsize=9)
ax.set_title('Jahresumsatz nach Produkt', fontsize=14, fontweight='bold')
ax.set_xlabel('Umsatz (Euro)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{x/1000:.0f}k'))
ax.set_xlim(0, df_p['umsatz'].max() * 1.15)
plt.tight_layout()
plt.savefig('output/umsatz_produkte.png', dpi=150, bbox_inches='tight')
plt.show()
````

### Aufgabe 2 — Zeitreihe mit Annotation

Erstelle ein Liniendiagramm des monatlichen Umsatzverlaufs mit:
- Markierung des Höchstwerts mit einem Pfeil (`ax.annotate`)
- Leicht eingefärbtem Bereich unter der Linie (`fill_between`)
- Gestrichelter Durchschnittslinie

````python
# LÖSUNG
import matplotlib.pyplot as plt

monate = ['Jan','Feb','Mär','Apr','Mai','Jun','Jul','Aug','Sep','Okt','Nov','Dez']
umsatz = [41000,38000,45000,52000,49000,61000,58000,55000,63000,71000,68000,85000]

avg     = sum(umsatz) / len(umsatz)
max_idx = umsatz.index(max(umsatz))

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(monate, umsatz, 'o-', color='#2196F3', linewidth=2.5, markersize=7)
ax.fill_between(monate, umsatz, alpha=0.15, color='#2196F3')
ax.axhline(avg, color='gray', linestyle='--', linewidth=1,
           label=f'Durchschnitt: {avg:,.0f} €')
ax.annotate(f'Höchstwert\n{max(umsatz):,} €',
            xy=(max_idx, max(umsatz)), xytext=(max_idx-2, max(umsatz)-8000),
            arrowprops=dict(arrowstyle='->', color='red'), color='red', fontsize=10)
ax.set_title('Monatlicher Umsatzverlauf', fontsize=14, fontweight='bold')
ax.set_ylabel('Umsatz (Euro)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

### Aufgabe 3 — Mini-Dashboard

Erstelle ein Dashboard mit 3 Diagrammen (1 Zeile, 3 Spalten):
- **Links**: Balkendiagramm — Umsatz pro Region
- **Mitte**: Kreisdiagramm — Marktanteil pro Kategorie
- **Rechts**: Liniendiagramm — Umsatztrend über 6 Monate

Vergib einen gemeinsamen Titel (`fig.suptitle`).

````python
# LÖSUNG
import matplotlib.pyplot as plt, matplotlib.ticker as mticker

regionen     = {'Nord': 68000, 'Süd': 54000, 'Ost': 41000, 'West': 37000}
kategorien   = {'Elektronik': 112000, 'Zubehör': 48000, 'Software': 40000}
trend_monate = ['Jan','Feb','Mär','Apr','Mai','Jun']
trend_umsatz = [28000, 32000, 29000, 38000, 41000, 36000]

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Vertriebs-Dashboard', fontsize=15, fontweight='bold')

axes[0].bar(regionen.keys(), regionen.values(), color='#2196F3')
axes[0].set_title('Umsatz pro Region', fontweight='bold')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{x/1000:.0f}k'))

axes[1].pie(kategorien.values(), labels=kategorien.keys(), autopct='%1.0f%%',
           startangle=90, colors=['#2196F3','#4CAF50','#FF9800'])
axes[1].set_title('Marktanteil', fontweight='bold')

axes[2].plot(trend_monate, trend_umsatz, 'o-', color='#4CAF50', linewidth=2.5, markersize=7)
axes[2].set_title('Umsatztrend', fontweight='bold')
axes[2].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'{x/1000:.0f}k'))
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('output/mini_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
````
