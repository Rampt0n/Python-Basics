# 3. Bedingungen (if / elif / else)

Mit **Bedingungen** kann dein Programm Entscheidungen treffen: "Wenn X, dann mache Y, sonst mache Z."

**Wichtig:** Python verwendet **Einrückung** (4 Leerzeichen oder 1 Tab), um zusammengehörige Code-Blöcke zu markieren!

Grundstruktur einer Bedingung:

```python
if BEDINGUNG:
    # Code wenn True
else:
    # Code wenn False
```

````python
umsatz = 5200
ziel = 5000

if umsatz >= ziel:
    print("Ziel erreicht! Gute Arbeit.")
else:
    print("Ziel noch nicht erreicht.")
````

````python
# elif = "else if" - für mehrere Bedingungen
bewertung = 72  # Punkte in einem Test (0-100)

if bewertung >= 90:
    note = "Sehr gut"
elif bewertung >= 75:
    note = "Gut"
elif bewertung >= 60:
    note = "Befriedigend"
elif bewertung >= 50:
    note = "Ausreichend"
else:
    note = "Nicht bestanden"

print(f"Ergebnis: {bewertung} Punkte - {note}")
````

### F-String-Formatierung

In F-Strings kann man nach einem **Doppelpunkt** `:` angeben, wie ein Wert formatiert werden soll:

```python
f"{wert:FORMAT}"
```

Die wichtigsten Formatierungsoptionen:

| Format | Bedeutung | Beispiel | Ergebnis |
|--------|-----------|----------|----------|
| `:.2f` | 2 Nachkommastellen | `f"{3.1:.2f}"` | `3.10` |
| `:.0f` | Keine Nachkommastellen | `f"{3.7:.0f}"` | `4` |
| `:.1f` | 1 Nachkommastelle | `f"{3.14:.1f}"` | `3.1` |
| `:,` | Tausendertrennzeichen | `f"{45000:,}"` | `45,000` |
| `:,.0f` | Tausender + keine Nachkomma | `f"{45000.5:,.0f}"` | `45,001` |
| `:<20` | Linksbündig, mind. 20 Zeichen breit | `f"{"Hallo":<20}"` | `Hallo···············` |
| `:>10` | Rechtsbündig, mind. 10 Zeichen breit | `f"{"Hallo":>10}"` | `·····Hallo` |

Die Punkte `···` stehen hier für Leerzeichen. Ausrichtung ist nützlich, um tabellarische Ausgaben zu erzeugen.

Formate lassen sich kombinieren: `:>10,.2f` bedeutet rechtsbündig, 10 Zeichen breit, Tausendertrenner, 2 Nachkommastellen.

````python
# Nützliche Funktionen für Zahlen-Listen
preise = [999.0, 49.90, 349.0, 79.90]

print(f"Summe: {sum(preise):.2f} Euro")
print(f"Minimum: {min(preise):.2f} Euro")
print(f"Maximum: {max(preise):.2f} Euro")
print(f"Durchschnitt: {sum(preise)/len(preise):.2f} Euro")
````

# 5. Schleifen (for / while)

Schleifen wiederholen Code-Blöcke — zum Beispiel um alle Elemente einer Liste zu verarbeiten.

## 5.1 Die for-Schleife

Die `for`-Schleife geht ein Element nach dem anderen durch eine Liste (oder andere iterable Objekte).

````python
produkte = ["Laptop", "Tastatur", "Monitor", "Headset"]

for produkt in produkte:
    print(f"Produkt: {produkt}")
````

`range()` erzeugt eine Zahlenfolge:

| Aufruf | Ergebnis |
|--------|----------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 (Schrittweite 2) |

````python

for i in range(1, 6):
    print(f"Durchlauf {i}")
````

````python
# enumerate() gibt Index und Wert gleichzeitig
produkte = ["Laptop", "Tastatur", "Monitor"]
preise = [999.0, 49.90, 349.0]

for i, produkt in enumerate(produkte):
    print(f"{i+1}. {produkt}: {preise[i]:.2f} Euro")
````

````python
# Praktisches Beispiel: Alle Preise mit 19% MwSt berechnen
netto_preise = [100.0, 250.0, 80.0, 430.0]
brutto_preise = []

for preis in netto_preise:
    brutto = preis * 1.19
    brutto_preise.append(brutto)

print("Netto:  ", netto_preise)
print("Brutto: ", brutto_preise)
````

## 5.2 Die while-Schleife

Die `while`-Schleife läuft so lange, wie eine Bedingung `True` ist.

````python
# while eignet sich gut, wenn du nicht weißt, wie viele Durchläufe nötig sind
kontostand = 1000
monat = 1

while kontostand > 0:
    kontostand -= 150  # -= ist kurz für: kontostand = kontostand - 150
    print(f"Monat {monat}: Kontostand {kontostand} Euro")
    monat += 1

print("Konto ist leer.")
````

# 6. Funktionen

Eine **Funktion** ist ein wiederverwendbarer Code-Block, dem man einen Namen gibt. Statt denselben Code mehrfach zu schreiben, definiert man eine Funktion und ruft sie auf.

**Prinzip**: DRY — "Don't Repeat Yourself"

```python
def funktionsname(parameter1, parameter2):
    # Code
    return ergebnis
```

Eine Funktion ohne Parameter und ohne Rückgabewert gibt nur etwas aus. `return` ist optional — führt eine Funktion nur eine Aktion durch (z.B. etwas ausgeben), braucht sie kein `return`.

````python
def begruessung_ausgeben():
    print("Willkommen im Python Workshop!")
    print("Heute lernst du die Grundlagen.")

# Funktion aufrufen:
begruessung_ausgeben()
````

````python
# Funktion mit Parametern
def begruessung_person(name, abteilung):
    print(f"Hallo {name} aus der Abteilung {abteilung}!")

begruessung_person("Anna", "IT")
begruessung_person("Klaus", "Vertrieb")
begruessung_person("Sara", "Buchhaltung")
````

**Parameter mit Standardwert (Default-Wert):**

```python
def funktion(pflicht_param, optionaler_param=standardwert)
```

Wird kein Wert übergeben, gilt automatisch der Standardwert.

````python

def mwst_berechnen(netto_preis, steuersatz=0.19):
    """Berechnet den Bruttopreis inkl. MwSt. (Standard-Steuersatz: 0.19 = 19%)."""
    brutto = netto_preis * (1 + steuersatz)
    return brutto

# Aufruf ohne steuersatz - Standardwert 0.19 wird verwendet
preis_laptop = mwst_berechnen(840.00)
print(f"Laptop inkl. 19% MwSt: {preis_laptop:.2f} Euro")

# Aufruf mit abweichendem Steuersatz (z.B. 7% für Lebensmittel)
preis_buch = mwst_berechnen(20.00, steuersatz=0.07)
print(f"Buch inkl. 7% MwSt: {preis_buch:.2f} Euro")
````

````python
# Funktion mit mehreren Rückgabewerten
def umsatz_analyse(umsaetze):
    """Analysiert eine Liste von Umsätzen."""
    gesamt = sum(umsaetze)
    durchschnitt = gesamt / len(umsaetze)
    maximum = max(umsaetze)
    minimum = min(umsaetze)
    return gesamt, durchschnitt, maximum, minimum

monatsumsaetze = [4200, 5100, 3800, 6200, 5500, 4900]

gesamt, avg, maxi, mini = umsatz_analyse(monatsumsaetze)
print(f"Gesamt:      {gesamt:.0f} Euro")
print(f"Durchschnitt: {avg:.0f} Euro")
print(f"Maximum:     {maxi:.0f} Euro")
print(f"Minimum:     {mini:.0f} Euro")
````

````python
# LÖSUNG
kunden_liste = [
    {"name": "Anna Schmidt",  "umsatz": 2450, "region": "Nord"},
    {"name": "Klaus Müller",  "umsatz": 890,  "region": "Süd"},
    {"name": "Sara Wagner",   "umsatz": 4100, "region": "Nord"},
    {"name": "Tom Becker",    "umsatz": 1200, "region": "Ost"},
]

# Aufgabe 1
# Gesamtumsatz per Schleife
gesamt = 0
for k in kunden_liste:
    gesamt += k["umsatz"]
print(f"Gesamtumsatz: {gesamt} Euro")

# Aufgabe 2
top_kunde = kunden_liste[0]
for kunde in kunden_liste:
    if kunde["umsatz"] > top_kunde["umsatz"]:
        top_kunde = kunde
print(f"Top-Kunde: {top_kunde['name']} ({top_kunde['umsatz']} Euro)")

# Aufgabe 3
umsatz_pro_region = {}
for kunde in kunden_liste:
    region = kunde["region"]
    if region in umsatz_pro_region:
        umsatz_pro_region[region] += kunde["umsatz"]
    else:
        umsatz_pro_region[region] = kunde["umsatz"]

print(f"Umsatz pro Region: {umsatz_pro_region}")
````

````python
# LÖSUNG
def bereinige_datensatz(datensatz):
    """Bereinigt einen einzelnen Verkaufs-Datensatz."""
    bereinigt = {}
    bereinigt["id"] = datensatz["id"]
    bereinigt["produkt"] = datensatz["produkt"].strip()
    bereinigt["produkt"] = bereinigt["produkt"].title()
    bereinigt["region"] = datensatz["region"].strip()
    bereinigt["region"] = bereinigt["region"].lower()
    
    # Preis umwandeln (ungültige Werte -> 0.0)
    if datensatz["preis"] in ("N/A", "", None, "-"):
        bereinigt["preis"] = 0.0
    else:
        bereinigt["preis"] = float(datensatz["preis"])
    
    # Menge umwandeln (leere Werte -> 0)
    if datensatz["menge"] in ("", None):
        bereinigt["menge"] = 0
    else:
        bereinigt["menge"] = int(datensatz["menge"])
    
    bereinigt["umsatz"] = bereinigt["preis"] * bereinigt["menge"]
    return bereinigt


# Alle Datensätze bereinigen
bereinigte_daten = []
for datensatz in rohdaten:
    bereinigte_daten.append(bereinige_datensatz(datensatz))

# Ergebnisse ausgeben
print(f"{'ID':<4} {'Produkt':<12} {'Preis':>8} {'Menge':>6} {'Umsatz':>10} {'Region':<8}")
print("-" * 55)
for d in bereinigte_daten:
    print(f"{d['id']:<4} {d['produkt']:<12} {d['preis']:>8.2f} {d['menge']:>6} {d['umsatz']:>10.2f} {d['region']:<8}")

# Zusammenfassung
# Gesamtumsatz per Schleife
gesamt_umsatz = 0
for d in bereinigte_daten:
    gesamt_umsatz += d["umsatz"]
print("-" * 55)
print(f"{'Gesamt-Umsatz':>32} {gesamt_umsatz:>10.2f}")

# Umsatz pro Region
print("\nUmsatz pro Region:")
regionen = {}
for d in bereinigte_daten:
    r = d["region"]
    regionen[r] = regionen.get(r, 0) + d["umsatz"]
for region, umsatz in sorted(regionen.items()):
    print(f"  {region.capitalize()}: {umsatz:.2f} Euro")
````

### Aufgabe 2 — Passwort-Prüfer

Schreibe eine Funktion `passwort_pruefen(pw)`, die ein Passwort bewertet:
- Mindestens 8 Zeichen → +1 Punkt
- Enthält eine Zahl → +1 Punkt
- Enthält einen Großbuchstaben → +1 Punkt

Rückgabe: `'schwach'` (0–1 Punkte), `'mittel'` (2), `'stark'` (3).

Teste mit: `'abc'`, `'passwort1'`, `'Passwort1'`

````python
# LÖSUNG
def passwort_pruefen(pw):
    punkte = 0
    if len(pw) >= 8:
        punkte += 1
    # Ziffern prüfen
    hat_ziffer = False
    for c in pw:
        if c.isdigit():
            hat_ziffer = True
            break
    if hat_ziffer:
        punkte += 1
    # Großbuchstaben prüfen
    hat_gross = False
    for c in pw:
        if c.isupper():
            hat_gross = True
            break
    if hat_gross:
        punkte += 1
    if punkte <= 1:
        return 'schwach'
    elif punkte == 2:
        return 'mittel'
    else:
        return 'stark'

for pw in ['abc', 'passwort1', 'Passwort1']:
    print(f"  '{pw}' -> {passwort_pruefen(pw)}")
````

### Aufgabe 3 — Häufigkeiten zählen

Zähle mit einem Dictionary, wie oft jede Stadt in der Liste vorkommt.
Gib danach die häufigste Stadt aus.

````python
# LÖSUNG
staedte = ['Berlin', 'Hamburg', 'Berlin', 'München', 'Hamburg',
           'Berlin', 'Köln', 'Hamburg', 'München', 'Berlin']

haeufigkeit = {}
for s in staedte:
    haeufigkeit[s] = haeufigkeit.get(s, 0) + 1

print(haeufigkeit)

# Häufigste Stadt per for-Schleife ermitteln
top_stadt = None
top_anzahl = 0
for stadt, anzahl in haeufigkeit.items():
    if anzahl > top_anzahl:
        top_anzahl = anzahl
        top_stadt = stadt
print(f'Häufigste Stadt: {top_stadt} ({top_anzahl}x)')
````

### Aufgabe 5 — Notenrechner

Schreibe eine Funktion `notenrechner(punkte_liste)`, die:
1. Den Durchschnitt berechnet
2. Die beste und schlechteste Note ermittelt
3. Einen zusammenfassenden Text zurückgibt

Teste mit: `[78, 92, 65, 88, 71, 95, 60]`

````python
# LÖSUNG
def notenrechner(punkte_liste):
    avg   = sum(punkte_liste) / len(punkte_liste)
    beste = max(punkte_liste)
    schlechteste = min(punkte_liste)
    return (f'Durchschnitt: {avg:.1f} | '
            f'Beste: {beste} | Schlechteste: {schlechteste}')

print(notenrechner([78, 92, 65, 88, 71, 95, 60]))
````
