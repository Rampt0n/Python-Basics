## Aufgabe 1: EXTRACT (Daten einlesen)

Importiere die notwendigen Bibliotheken und lade die bereitgestellte CSV-Datei als DataFrame.

#### Benötigte Befehle:
* `import pandas as pd`: Lädt die Bibliothek für Datenanalyse.
* `pd.read_csv(...)`: Der Befehl von Pandas, um Daten in eine Tabelle (`df_roh`) einzulesen.

```python
import pandas as pd

# Daten einlesen
df_roh = pd.read_csv("bsp_daten/rohdaten.csv")
```

## Aufgabe 2: Sichten der Daten

Verschaffe dir einen Überblick über die eingelese Tabelle,  
die jetzt als Pandas Dataframe vorliegt.

**Dein Spickzettel:**
* `df.dtypes`: Zeigt den erkannten Datentyp jeder Spalte an.

```python
print("\n--- Analyse der Datentypen ---")
# Zeigt, dass die Beitrags-Spalte als Text (object) erkannt wurde
print(df_roh.dtypes)

print("--- Die eingelesenen Rohdaten ---")
display(df_roh)
```


### Aufgabe 3:
Bereinige alle Texte so, dass diese einheitlich und weiterverarbeitbar sind.

1. Entferne führende und folgende Leerzeichen aus `Versicherungsart`. 

Benötigte Befehle: 
* `.str.title()`: Formatiert den String so, sodass der erste Buchstabe großgeschrieben wird und der Rest klein.  


```python
df_roh["Versicherungsart"] = df_roh["Versicherungsart"].str.title() 
df_roh
```

## Aufgabe 4: TRANSFORM (Teil 2: Typenwandlung und neue Metriken)

**Deine Aufgabe:** Wandle den nun sauberen Text in echte Zahlen um, lösche die alte Text-Spalte und berechne den Jahresumsatz.
Erstelle vor dem Arbeiten mit dem Dataframe eine Kopie und nenne diesen  `df_transformed`

**Dein Spickzettel:**
* `.copy()` : Erstellt eine Kopie eines PandasDataframes.
* `.astype(int)`: Wandelt Werte in den angegebenen Datentyp um.
* `df['A'] * df['B']`: Multipliziert Spalten miteinander.

```python
df_transformed = df_roh.copy()  # Erstellen einer Kopie der Rohdaten für die Transformation 
df_transformed["Aktive_Vertraege"] = df_transformed["Aktive_Vertraege"].astype(int)
df_transformed["Jahresumsatz"] = df_transformed["Monatsbeitrag_Text"]* 12 * df_transformed["Aktive_Vertraege"]
df_transformed
```

## Aufgabe 5: **VISUALISIERUNG**

**Deine Aufgabe:**  
Sortiere die bereinigten Daten aufsteigend nach dem Umsatz und erstelle anschließend das horizontale Balkendiagramm. Die optischen Formatierungen werden automatisch über die `df.plot()`‑Funktion übernommen.

**Dein Spickzettel:**
* `.sort_values(by='...', ascending=True)`: Sortiert die Tabelle.
* `.plot(kind='barh', x='...', y='...')`: Erstellt ein horizontales Balkendiagramm aus einem DataFrame.

```python
# Daten sortieren
df_clean = df_transformed.copy()
df_clean = df_clean.sort_values('Jahresumsatz')

# Minimaler Plot
df_clean.plot(
    kind='barh',
    x='Versicherungsart',
    y='Jahresumsatz'
)
```
