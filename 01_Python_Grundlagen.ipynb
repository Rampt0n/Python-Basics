## 2.1 Zahlen

Python kennt zwei Zahlentypen:
- `int` (Integer) - ganze Zahlen: `1`, `42`, `-7`
- `float` (Fließkommazahl) - Dezimalzahlen: `3.14`, `-0.5`, `1.0`

````python
anzahl_kunden = 1500          # int
umsatz = 48750.99             # float
wachstum_prozent = 12.5       # float

print(anzahl_kunden)
print(umsatz)
print(wachstum_prozent)
````

````python
# type() zeigt den Datentyp einer Variable an
print(type(anzahl_kunden))
print(type(umsatz))
````

### Rechenoperationen mit Zahlen

| Operator | Bedeutung | Beispiel |
|----------|-----------|----------|
| `+` | Addition | `3 + 2 = 5` |
| `-` | Subtraktion | `10 - 4 = 6` |
| `*` | Multiplikation | `3 * 4 = 12` |
| `/` | Division | `7 / 2 = 3.5` |
| `//` | Ganzzahldivision | `7 // 2 = 3` |
| `%` | Modulo (Rest) | `7 % 2 = 1` |

````python
preis_pro_einheit = 29.90
menge = 100

gesamtpreis = preis_pro_einheit * menge
print("Gesamtpreis:", gesamtpreis)

rabatt = gesamtpreis * 0.10   # 10% Rabatt
endpreis = gesamtpreis - rabatt
print("Endpreis nach 10% Rabatt:", endpreis)
````

## 2.2 Text (Strings)

Texte heißen in Python **Strings** (Typ: `str`). Sie werden in einfache `'...'` oder doppelte `"..."` Anführungszeichen eingeschlossen.

````python
vorname = "Maria"
nachname = 'Müller'
abteilung = "Vertrieb"

print(vorname)
print(type(vorname))
````

````python
# Strings zusammenfügen (Konkatenation) mit +
voller_name = vorname + " " + nachname
print(voller_name)
````

**F-Strings** sind die modernste und lesbarste Art, Variablen in Text einzubauen. Das `f` vor dem Anführungszeichen aktiviert den F-String — Variablen kommen in geschweifte Klammern `{}`.

````python

begruessung = f"Willkommen, {vorname} {nachname} aus der Abteilung {abteilung}!"
print(begruessung)
````

````python
# Nützliche String-Methoden
text = "  Anna Schmidt  "

bereinigt = text.strip()         # Leerzeichen am Rand entfernen
print(bereinigt)
print(bereinigt.upper())         # Großbuchstaben
print(bereinigt.lower())         # Kleinbuchstaben
print(bereinigt.title())         # Titel-Format (Erster Buchstabe groß )
print(len(bereinigt))            # Länge des Strings

satz = "Python ist toll"
print(satz.replace("toll", "super"))   # Ersetzen
print(satz.split(" "))                 # Aufteilen in Liste
````

## 2.3 Wahrheitswerte (Boolean)

Ein **Boolean** (`bool`) kann nur zwei Werte annehmen: `True` oder `False`. Sie entstehen oft durch Vergleiche.

````python
ist_aktiv = True
hat_premium = False

print(ist_aktiv)
print(type(ist_aktiv))
````

Vergleichsoperatoren erzeugen `True` oder `False`:

| Operator | Bedeutung |
|----------|-----------|
| `==` | gleich (nicht `=` wie Zuweisung!) |
| `!=` | ungleich |
| `>` | größer als |
| `<` | kleiner als |
| `>=` | größer oder gleich |
| `<=` | kleiner oder gleich |

````python

umsatz = 5000
ziel = 4000

print(umsatz > ziel)     # True
print(umsatz == ziel)    # False
print(umsatz != ziel)    # True
````

## 2.4 Datentypen umwandeln

Manchmal müssen Datentypen umgewandelt werden - z.B. wenn man eine Zahl als Text einliest (typisch bei CSV-Daten!).

````python
# Aus einer CSV-Datei kommen Zahlen oft als Text an:
umsatz_text = "12500.50"   # Das ist ein String, keine Zahl!

# Umwandlung:
umsatz_zahl = float(umsatz_text)
print(umsatz_zahl + 1000)   # Jetzt kann man rechnen

# Umgekehrt: Zahl zu Text
zahl = 42
text = str(zahl)
print("Ergebnis: " + text)  # Konkatenation klappt jetzt

# int() wandelt in ganze Zahl um (schneidet Nachkommastellen ab)
print(int(3.9))   # Ergibt 3, nicht 4!
````

# 3. Bedingungen (if / elif / else)

Mit **Bedingungen** kann dein Programm Entscheidungen treffen: "Wenn X, dann mache Y, sonst mache Z."

**Wichtig:** Python verwendet **Einrückung** (4 Leerzeichen oder 1 Tab), um zusammengehörige Code-Blöcke zu markieren!

Grundstruktur einer Bedingung:

```python
if BEDINGUNG:
    # Code wenn True
else:
    # Code wenn False
```

````python
umsatz = 5200
ziel = 5000

if umsatz >= ziel:
    print("Ziel erreicht! Gute Arbeit.")
else:
    print("Ziel noch nicht erreicht.")
````

````python
# elif = "else if" - für mehrere Bedingungen
bewertung = 72  # Punkte in einem Test (0-100)

if bewertung >= 90:
    note = "Sehr gut"
elif bewertung >= 75:
    note = "Gut"
elif bewertung >= 60:
    note = "Befriedigend"
elif bewertung >= 50:
    note = "Ausreichend"
else:
    note = "Nicht bestanden"

print(f"Ergebnis: {bewertung} Punkte - {note}")
````

````python
# Logische Operatoren: and, or, not
alter = 28
hat_führerschein = True

if alter >= 18 and hat_führerschein:
    print("Darf Auto fahren")
else:
    print("Darf nicht Auto fahren")
````

# 4. Listen

Eine **Liste** speichert mehrere Werte in einer bestimmten Reihenfolge. Listen werden mit eckigen Klammern `[]` erstellt.

Listen sind der erste Schritt hin zu Datentabellen - und damit zu Pandas und ETL-Pipelines!

````python
produkte = ["Laptop", "Maus", "Tastatur", "Monitor", "Headset"]
preise = [999.0, 29.90, 49.90, 349.0, 79.90]

print(produkte)
print(f"Anzahl Produkte: {len(produkte)}")
````

````python
# Auf einzelne Elemente zugreifen: Index beginnt bei 0!
print(produkte[0])   # Erstes Element: "Laptop"
print(produkte[1])   # Zweites Element: "Maus"
print(produkte[-1])  # Letztes Element: "Headset"

# Slicing: Teilbereich einer Liste
print(produkte[1:3])  # Index 1 bis 2 (3 ist ausgeschlossen)
````

````python
# Listen verändern
produkte.append("Webcam")       # Element hinzufügen
print(produkte)

produkte.remove("Maus")         # Element entfernen
print(produkte)

produkte[0] = "Gaming-Laptop"   # Element ersetzen
print(produkte)
````

````python
# 'in' prüft, ob ein Element in einer Liste vorkommt
vip_status = ["Premium", "Gold", "Platin"]

status = "Gold"
if status in vip_status:
    print("Ist VIP-Kunde")

status2 = "Standard"
if status2 not in vip_status:
    print("Kein VIP-Kunde")
````

### Exkurs: F-String-Formatierung

In F-Strings kann man nach einem **Doppelpunkt** `:` angeben, wie ein Wert formatiert werden soll:

```python
f"{wert:FORMAT}"
```

Die wichtigsten Formatierungsoptionen:

| Format | Bedeutung | Beispiel | Ergebnis |
|--------|-----------|----------|----------|
| `:.2f` | 2 Nachkommastellen | `f"{3.1:.2f}"` | `3.10` |
| `:.0f` | Keine Nachkommastellen | `f"{3.7:.0f}"` | `4` |
| `:.1f` | 1 Nachkommastelle | `f"{3.14:.1f}"` | `3.1` |
| `:,` | Tausendertrennzeichen | `f"{45000:,}"` | `45,000` |
| `:,.0f` | Tausender + keine Nachkomma | `f"{45000.5:,.0f}"` | `45,001` |
| `:<20` | Linksbündig, mind. 20 Zeichen breit | `f"{"Hallo":<20}"` | `Hallo···············` |
| `:>10` | Rechtsbündig, mind. 10 Zeichen breit | `f"{"Hallo":>10}"` | `·····Hallo` |

Die Punkte `···` stehen hier für Leerzeichen. Ausrichtung ist nützlich, um tabellarische Ausgaben zu erzeugen.

Formate lassen sich kombinieren: `:>10,.2f` bedeutet rechtsbündig, 10 Zeichen breit, Tausendertrenner, 2 Nachkommastellen.

````python
# Nützliche Funktionen für Zahlen-Listen
preise = [999.0, 49.90, 349.0, 79.90]

print(f"Summe: {sum(preise):.2f} Euro")
print(f"Minimum: {min(preise):.2f} Euro")
print(f"Maximum: {max(preise):.2f} Euro")
print(f"Durchschnitt: {sum(preise)/len(preise):.2f} Euro")
````

# 5. Schleifen (for / while)

Schleifen wiederholen Code-Blöcke - zum Beispiel um alle Elemente einer Liste zu verarbeiten.

## 5.1 Die for-Schleife

Die `for`-Schleife geht ein Element nach dem anderen durch eine Liste (oder andere iterable Objekte).

````python
produkte = ["Laptop", "Tastatur", "Monitor", "Headset"]

for produkt in produkte:
    print(f"Produkt: {produkt}")
````

`range()` erzeugt eine Zahlenfolge:

| Aufruf | Ergebnis |
|--------|----------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 (Schrittweite 2) |

````python

for i in range(1, 6):
    print(f"Durchlauf {i}")
````

````python
# enumerate() gibt Index und Wert gleichzeitig
produkte = ["Laptop", "Tastatur", "Monitor"]
preise = [999.0, 49.90, 349.0]

for i, produkt in enumerate(produkte):
    print(f"{i+1}. {produkt}: {preise[i]:.2f} Euro")
````

````python
# Praktisches Beispiel: Alle Preise mit 19% MwSt berechnen
netto_preise = [100.0, 250.0, 80.0, 430.0]
brutto_preise = []

for preis in netto_preise:
    brutto = preis * 1.19
    brutto_preise.append(brutto)

print("Netto:  ", netto_preise)
print("Brutto: ", brutto_preise)
````

## 5.2 Die while-Schleife

Die `while`-Schleife läuft so lange, wie eine Bedingung `True` ist.

````python
# while eignet sich gut, wenn du nicht weißt, wie viele Durchläufe nötig sind
kontostand = 1000
monat = 1

while kontostand > 0:
    kontostand -= 150  # -= ist kurz für: kontostand = kontostand - 150
    print(f"Monat {monat}: Kontostand {kontostand} Euro")
    monat += 1

print("Konto ist leer.")
````

# 6. Funktionen

Eine **Funktion** ist ein wiederverwendbarer Code-Block, dem man einen Namen gibt. Statt denselben Code mehrfach zu schreiben, definiert man eine Funktion und ruft sie auf.

**Prinzip**: DRY - "Don't Repeat Yourself"

```python
def funktionsname(parameter1, parameter2):
    # Code
    return ergebnis
```

Eine Funktion ohne Parameter und ohne Rückgabewert gibt nur etwas aus. `return` ist optional — führt eine Funktion nur eine Aktion durch (z.B. etwas ausgeben), braucht sie kein `return`.

````python
def begruessung_ausgeben():
    print("Willkommen im Python Workshop!")
    print("Heute lernst du die Grundlagen.")

# Funktion aufrufen:
begruessung_ausgeben()
````

````python
# Funktion mit Parametern
def begruessung_person(name, abteilung):
    print(f"Hallo {name} aus der Abteilung {abteilung}!")

begruessung_person("Anna", "IT")
begruessung_person("Klaus", "Vertrieb")
begruessung_person("Sara", "Buchhaltung")
````

**Parameter mit Standardwert (Default-Wert):**

```python
def funktion(pflicht_param, optionaler_param=standardwert)
```

Wird kein Wert übergeben, gilt automatisch der Standardwert.

````python

def mwst_berechnen(netto_preis, steuersatz=0.19):
    """Berechnet den Bruttopreis inkl. MwSt. (Standard-Steuersatz: 0.19 = 19%)."""
    brutto = netto_preis * (1 + steuersatz)
    return brutto

# Aufruf ohne steuersatz - Standardwert 0.19 wird verwendet
preis_laptop = mwst_berechnen(840.00)
print(f"Laptop inkl. 19% MwSt: {preis_laptop:.2f} Euro")

# Aufruf mit abweichendem Steuersatz (z.B. 7% für Lebensmittel)
preis_buch = mwst_berechnen(20.00, steuersatz=0.07)
print(f"Buch inkl. 7% MwSt: {preis_buch:.2f} Euro")
````

````python
# Funktion mit mehreren Rückgabewerten
def umsatz_analyse(umsaetze):
    """Analysiert eine Liste von Umsätzen."""
    gesamt = sum(umsaetze)
    durchschnitt = gesamt / len(umsaetze)
    maximum = max(umsaetze)
    minimum = min(umsaetze)
    return gesamt, durchschnitt, maximum, minimum

monatsumsaetze = [4200, 5100, 3800, 6200, 5500, 4900]

gesamt, avg, maxi, mini = umsatz_analyse(monatsumsaetze)
print(f"Gesamt:      {gesamt:.0f} Euro")
print(f"Durchschnitt: {avg:.0f} Euro")
print(f"Maximum:     {maxi:.0f} Euro")
print(f"Minimum:     {mini:.0f} Euro")
````

# 7. Dictionaries

Ein **Dictionary** (`dict`) speichert Daten als Schlüssel-Wert-Paare. Statt eines Index (wie bei Listen) greift man über einen **Schlüssel** (Key) auf Werte zu.

Ein Dictionary ist wie eine Akte: Jede Seite (Wert) hat eine Beschriftung (Schlüssel).

Dictionaries sind besonders wichtig beim Umgang mit JSON-Daten!

````python
# Dictionary erstellen mit geschweiften Klammern {}
kunde = {
    "name": "Anna Schmidt",
    "alter": 34,
    "stadt": "Hamburg",
    "premium": True,
    "jahresumsatz": 2450.80
}

print(kunde)
````

````python
# Auf Werte zugreifen
print(kunde["name"])    # Über den Schlüssel in eckigen Klammern
print(kunde["alter"])
````

````python
# Dictionary verändern
kunde["email"] = "a.schmidt@beispiel.de"  # Neuen Schlüssel hinzufügen
kunde["alter"] = 35                        # Wert ändern
del kunde["premium"]                       # Schlüssel löschen

print(kunde)
````

````python
# Über ein Dictionary iterieren
print("Alle Schlüssel:", list(kunde.keys()))
print("Alle Werte:  ", list(kunde.values()))
print()

# items() gibt Schlüssel-Wert-Paare zurück
for schluessel, wert in kunde.items():
    print(f"  {schluessel}: {wert}")
````

````python
# Liste von Dictionaries - so kommen Daten oft aus Datenbanken oder APIs
kunden_liste = [
    {"name": "Anna Schmidt",  "umsatz": 2450, "region": "Nord"},
    {"name": "Klaus Müller",  "umsatz": 890,  "region": "Süd"},
    {"name": "Sara Wagner",   "umsatz": 4100, "region": "Nord"},
    {"name": "Tom Becker",    "umsatz": 1200, "region": "Ost"},
]

# Alle Kunden aus der Region Nord
print("Nord-Kunden:")
for kunde in kunden_liste:
    if kunde["region"] == "Nord":
        print(f"  {kunde['name']}: {kunde['umsatz']} Euro")
````
