# 1. Balkendiagramm (Bar Chart)

**Wann?** Kategorien vergleichen (z.B. Umsatz pro Region)

`FuncFormatter` passt Achsenbeschriftungen an: Die Hilfsfunktion bekommt den Achsenwert als `x` übergeben und gibt einen formatierten Text zurück — z.B. `fmt_euro(45000, _)` → `"45.000 €"`. Die Funktionen `fmt_euro`, `fmt_kilo_euro` und `fmt_kilo` sind in der Setup-Zelle definiert.

````python
# Daten aggregieren
umsatz_region = df.groupby("region")["umsatz"].sum().sort_values(ascending=False)

# Diagramm erstellen
fig, ax = plt.subplots()  # fig = Zeichenfläche, ax = Achsensystem

bars = ax.bar(
    umsatz_region.index,     # x-Achse: Kategorien
    umsatz_region.values,    # y-Achse: Werte
    color=["#2196F3", "#4CAF50", "#FF9800"],  # Farben
    edgecolor="white",
    linewidth=0.8
)

# Werte über den Balken
for bar in bars:
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # x-Position: Mitte des Balkens
        bar.get_height() + 200,              # y-Position: leicht über dem Balken
        f"{bar.get_height():,.0f} €",        # Text
        ha="center", va="bottom", fontsize=10
    )

# Beschriftungen und Formatierung
ax.set_title("Gesamtumsatz nach Region (Q1 2024)", fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("Region")
ax.set_ylabel("Umsatz (Euro)")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(fmt_euro))
ax.set_ylim(0, umsatz_region.max() * 1.15)

plt.tight_layout()
plt.savefig("output/umsatz_region.png", dpi=150, bbox_inches="tight")
plt.show()
print("Gespeichert: output/umsatz_region.png")
````

# 2. Gruppierte Balken und horizontale Balken

````python
# Gruppiertes Balkendiagramm: Umsatz nach Region UND Kategorie
pivot_daten = df.groupby(["region", "kategorie"])["umsatz"].sum().unstack()

fig, ax = plt.subplots()
pivot_daten.plot(kind="bar", ax=ax, width=0.7)

ax.set_title("Umsatz nach Region und Kategorie", fontsize=14, fontweight="bold")
ax.set_xlabel("Region")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(title="Kategorie")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)  # Beschriftung horizontal
ax.yaxis.set_major_formatter(mticker.FuncFormatter(fmt_kilo_euro))

plt.tight_layout()
plt.show()
````

````python
# Gestapelte Balken (Stacked Bar) - für Anteile
fig, ax = plt.subplots()
pivot_daten.plot(kind="bar", stacked=True, ax=ax, width=0.6)

ax.set_title("Umsatz-Zusammensetzung nach Region", fontsize=14, fontweight="bold")
ax.set_xlabel("Region")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(title="Kategorie", loc="upper right")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

plt.tight_layout()
plt.show()
````

# 3. Liniendiagramm (Line Chart)

**Wann?** Entwicklung über Zeit zeigen

````python
# Umsatz-Trend pro Monat und Region
monat_region = df.groupby(["monat", "region"])["umsatz"].sum().reset_index()

# Monatsreihenfolge festlegen
monat_reihenfolge = ["Jan", "Feb", "Mär"]
monat_region["monat"] = pd.Categorical(monat_region["monat"], categories=monat_reihenfolge, ordered=True)
monat_region = monat_region.sort_values("monat")

fig, ax = plt.subplots()

farben = {"nord": "#2196F3", "süd": "#4CAF50", "ost": "#FF9800"}
marker = {"nord": "o", "süd": "s", "ost": "^"}

for region in ["nord", "süd", "ost"]:
    region_df = monat_region[monat_region["region"] == region]
    ax.plot(
        region_df["monat"],
        region_df["umsatz"],
        marker=marker[region],
        linewidth=2,
        markersize=8,
        label=f"Region {region.capitalize()}",
        color=farben[region]
    )

ax.set_title("Umsatz-Entwicklung nach Region (Q1 2024)", fontsize=14, fontweight="bold")
ax.set_xlabel("Monat")
ax.set_ylabel("Umsatz (Euro)")
ax.legend()
ax.yaxis.set_major_formatter(mticker.FuncFormatter(fmt_kilo_euro))
ax.grid(True, alpha=0.4)

plt.tight_layout()
plt.show()
````

# 4. Kreisdiagramm (Pie Chart)

**Wann?** Anteile am Ganzen zeigen (max. 5–6 Segmente)

````python
umsatz_kat = df.groupby("kategorie")["umsatz"].sum()

fig, ax = plt.subplots(figsize=(8, 6))

wedges, texts, autotexts = ax.pie(
    umsatz_kat.values,
    labels=umsatz_kat.index,
    autopct="%1.1f%%",          # Prozentangabe
    startangle=90,               # Startwinkel
    explode=[0.05] * len(umsatz_kat),  # Segmente leicht herausziehen
    colors=["#2196F3", "#4CAF50"],
    shadow=True
)

# Schriftgröße anpassen
for text in autotexts:
    text.set_fontsize(12)
    text.set_fontweight("bold")

ax.set_title("Umsatzanteil nach Kategorie", fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
````

# 5. Streudiagramm (Scatter Plot)

**Wann?** Zusammenhänge zwischen zwei Variablen untersuchen

````python
fig, ax = plt.subplots()

# Farben nach Kategorie
farben_map = {"Elektronik": "#2196F3", "Zubehör": "#FF9800"}

for kat, gruppe in df.groupby("kategorie"):
    ax.scatter(
        gruppe["bestellungen"],
        gruppe["umsatz"],
        label=kat,
        color=farben_map[kat],
        s=80,            # Punktgröße
        alpha=0.8,       # Transparenz
        edgecolors="white",
        linewidth=0.5
    )

ax.set_title("Bestellungen vs. Umsatz", fontsize=14, fontweight="bold")
ax.set_xlabel("Anzahl Bestellungen")
ax.set_ylabel("Umsatz (Euro)")
ax.legend(title="Kategorie")
ax.grid(True, alpha=0.3)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(fmt_kilo_euro))

plt.tight_layout()
plt.show()
````

# 6. Histogramm und Box Plot

**Wann?** Verteilung einer Variablen verstehen

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Histogramm
ax1.hist(df["umsatz"], bins=10, color="#2196F3", edgecolor="white", linewidth=0.8)
ax1.axvline(df["umsatz"].mean(), color="red", linestyle="--", label=f"Mittelwert: {df['umsatz'].mean():,.0f} €")
ax1.axvline(df["umsatz"].median(), color="orange", linestyle="--", label=f"Median: {df['umsatz'].median():,.0f} €")
ax1.set_title("Verteilung der Umsätze", fontsize=13, fontweight="bold")
ax1.set_xlabel("Umsatz (Euro)")
ax1.set_ylabel("Häufigkeit")
ax1.legend()

# Box Plot: Zeigt Median, Quartile, Ausreißer
daten_nach_region = []
for r in ["nord", "süd", "ost"]:
    daten_nach_region.append(df[df["region"] == r]["umsatz"].values)
bp = ax2.boxplot(
    daten_nach_region,
    labels=["Nord", "Süd", "Ost"],
    patch_artist=True,
    boxprops=dict(facecolor="#E3F2FD")
)
ax2.set_title("Umsatz-Verteilung nach Region", fontsize=13, fontweight="bold")
ax2.set_ylabel("Umsatz (Euro)")
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(fmt_kilo_euro))

plt.tight_layout()
plt.show()
````

# 7. Mehrere Diagramme in einem Layout (Subplots)

````python
# Dashboard-ähnliches Layout mit 4 Diagrammen
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Verkaufs-Dashboard Q1 2024", fontsize=16, fontweight="bold", y=1.02)

# --- Diagramm 1 (oben links): Umsatz nach Region ---
ax = axes[0, 0]
region_sum = df.groupby("region")["umsatz"].sum().sort_values(ascending=True)
ax.barh(region_sum.index, region_sum.values, color="#2196F3")
ax.set_title("Gesamtumsatz nach Region", fontweight="bold")
ax.set_xlabel("Euro")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(fmt_kilo))
for i, v in enumerate(region_sum.values):
    ax.text(v + 300, i, f"{v:,.0f} €", va="center", fontsize=9)

# --- Diagramm 2 (oben rechts): Monatstrend ---
ax = axes[0, 1]
monat_sum = df.groupby("monat")["umsatz"].sum()
monat_sum = monat_sum.reindex(["Jan", "Feb", "Mär"])
ax.plot(monat_sum.index, monat_sum.values, "o-", color="#4CAF50", linewidth=2.5, markersize=8)
ax.fill_between(monat_sum.index, monat_sum.values, alpha=0.2, color="#4CAF50")
ax.set_title("Umsatz-Trend", fontweight="bold")
ax.set_ylabel("Euro")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(fmt_kilo_euro))
ax.grid(True, alpha=0.3)

# --- Diagramm 3 (unten links): Kategorie-Anteile ---
ax = axes[1, 0]
kat_sum = df.groupby("kategorie")["umsatz"].sum()
ax.pie(kat_sum.values, labels=kat_sum.index, autopct="%1.0f%%",
       colors=["#2196F3", "#FF9800"], startangle=90)
ax.set_title("Umsatz nach Kategorie", fontweight="bold")

# --- Diagramm 4 (unten rechts): Bestellungen nach Monat und Region ---
ax = axes[1, 1]
pivot_best = df.groupby(["monat", "region"])["bestellungen"].sum().unstack()
pivot_best = pivot_best.reindex(["Jan", "Feb", "Mär"])
pivot_best.plot(kind="bar", ax=ax, width=0.7)
ax.set_title("Bestellungen nach Monat und Region", fontweight="bold")
ax.set_xlabel("")
ax.set_ylabel("Anzahl")
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
ax.legend(title="Region")

plt.tight_layout()
plt.savefig("output/dashboard.png", dpi=150, bbox_inches="tight")
plt.show()
print("Dashboard gespeichert: output/dashboard.png")
````

# 8. Seaborn - Statistische Diagramme

Seaborn vereinfacht viele häufige Visualisierungsaufgaben erheblich.

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Heatmap: Umsatz nach Monat und Region
pivot = df.groupby(["monat", "region"])["umsatz"].sum().unstack()
pivot = pivot.reindex(["Jan", "Feb", "Mär"])

sns.heatmap(
    pivot,
    ax=axes[0],
    annot=True,             # Werte in den Zellen anzeigen
    fmt=",.0f",             # Zahlenformat
    cmap="Blues",           # Farbschema
    linewidths=0.5
)
axes[0].set_title("Umsatz-Heatmap (Monat × Region)", fontweight="bold")

# Violinplot: Verteilung nach Kategorie
sns.violinplot(
    data=df, x="kategorie", y="umsatz",
    ax=axes[1],
    palette="muted", inner="box"
)
axes[1].set_title("Umsatz-Verteilung nach Kategorie", fontweight="bold")
axes[1].set_ylabel("Umsatz (Euro)")

plt.tight_layout()
plt.show()
````

# 9. Interaktive Diagramme mit plotly

Plotly-Diagramme sind interaktiv: Zoom, Hover, Pan - ideal für Reports.

````python
import plotly.express as px

# Interaktives Balkendiagramm
df_agg = df.groupby(["monat", "region"])["umsatz"].sum().reset_index()
df_agg["monat"] = pd.Categorical(df_agg["monat"], categories=["Jan", "Feb", "Mär"], ordered=True)
df_agg = df_agg.sort_values("monat")

fig = px.bar(
    df_agg,
    x="monat",
    y="umsatz",
    color="region",
    barmode="group",
    title="Umsatz nach Monat und Region (interaktiv)",
    labels={"umsatz": "Umsatz (Euro)", "monat": "Monat", "region": "Region"},
    color_discrete_map={"nord": "#2196F3", "süd": "#4CAF50", "ost": "#FF9800"}
)
fig.update_layout(legend_title="Region")
fig.show()
````
